```python
import cv2
from matplotlib import pyplot as plt
from pathlib import Path

# Tahmin sonuçlarının saklandığı dizin
output_dir = Path('/content/drive/MyDrive/Colab Notebooks/INBREAST_YoloV7/yolov7/runs/detect/exp')

# Sonuçları görselleştirmek için fonksiyon
def visualize_predictions(image_path):
    img = cv2.imread(str(image_path))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.figure(figsize=(10, 10))
    plt.imshow(img)
    plt.title(f'Predictions for {image_path.name}')
    plt.axis('off')
    plt.show()

# Tüm sonuçları görselleştirme
for img_path in output_dir.glob('*.jpg'):  # Resim dosya formatınıza göre düzenleyin
    visualize_predictions(img_path)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

# results.txt dosyasının yolu
txt_path = "/content/drive/MyDrive/Colab Notebooks/INBREAST_YoloV7/yolov7/runs/train/exp/results.txt"

# results.txt dosyasını okuyun
df = pd.read_csv(txt_path, delim_whitespace=True, header=None)

# Sütun adlarını belirleyin
df.columns = ['epoch', 'gpu_mem', 'box_loss', 'cls_loss', 'dfl_loss', 'total_loss', 'img_size',
              'val_precision', 'val_recall', 'val_mAP50', 'val_mAP50_95', 'test_precision',
              'test_recall', 'test_mAP50', 'test_mAP50_95']

# Gerekli sütunları alın
epochs = df['epoch']
train_loss = df['box_loss'] + df['cls_loss'] + df['dfl_loss']
val_loss = df['total_loss']
val_mAP50 = df['val_mAP50']

# Eğitim ve doğrulama kayıplarını çizdirin
plt.figure(figsize=(10, 6))
plt.plot(epochs, val_loss, 'r-', label='Doğrulama Kaybı')
plt.xlabel('Epochs')
plt.ylabel('Kayıp')
plt.title('YoloV7 Doğrulama Kaybı')
plt.legend()
plt.grid(True)
plt.gca().xaxis.set_major_locator(MultipleLocator(7))  # X eksenini 7 periyot arayla etiketleme
plt.show()


```

```python
# mAP@0.5 değerlerini çizdirin
plt.figure(figsize=(10, 6))
plt.plot(epochs, val_mAP50, 'g-', label='mAP@0.5')
plt.xlabel('Epochs')
plt.ylabel('mAP@0.5')
plt.title('YoloV7 mAP@0.5 Değeri')
plt.legend()
plt.grid(True)
plt.gca().xaxis.set_major_locator(MultipleLocator(7))  # X eksenini 7 periyot arayla etiketleme
plt.show()
```
